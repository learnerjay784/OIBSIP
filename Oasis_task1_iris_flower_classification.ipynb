Task : Iris flower has three species; setosa, versicolor, and virginica, which   differs according to their
measurements. Now assume that you have the measurements of the iris flowers according to
their species, and here your task is to train a machine learning model that can learn from the
measurements of the iris species and classify them.


Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier
```

Loading the dataset

```python
dataset = load_iris()
```

Summarising the dataset

```python
print(dataset.data)
print(dataset.target)

print(dataset.data.shape)
```

Conveting the dataset to pandas dataFrame

```python
X = pd.DataFrame(dataset.data, columns = dataset.feature_names)
print(X)
```

Statistical description of the data

```python
X.describe()
```

```python
Y = dataset.target
print(Y)
```

Labels:

Iris- setosa : 0 

Iris- vercolor : 1 

Iris- virginica : 2




Splitting the dataset into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size= 0.25, random_state=1)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Visualisation of the model

```python
accuracy = []

for i in range(1, 10):
  model = DecisionTreeClassifier(max_depth = i, random_state=0)
  model.fit(X_train, Y_train)
  pred = model.predict(X_test)
  score = accuracy_score(Y_test, pred)
  accuracy.append(score)

plt.figure(figsize = (12,6))
plt.plot(range(1,10), accuracy, color = 'red', linestyle = 'dashed',
         marker = 'o',mfc = 'blue', ms = 10)
plt.title('Finding best Max_Depth',size = 20)
plt.xlabel('Pred',size = 20)
plt.ylabel('Score',size = 20)


```

Training the DecisionTree Classifier

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth = 3, random_state = 0)
model.fit(X_train,Y_train)
```

Prediction

```python
Y_pred = model.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

Model Evaluation : Accuracy Score

```python
print('Accuracy of the model: {0}%'.format(accuracy_score(Y_test, Y_pred)*100))
```
