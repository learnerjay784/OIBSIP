**Task:** Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, so analyzing the unemployment rate can be a good data science project.



Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Loading the dataset

```python
data = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
```

Printing first five rows

```python
data.head()
```

Number of rows and columns 

```python
data.shape
```

Statistical analysis of dataset

```python
data.describe()
```

Renaming the column names

```python
data = data.rename(columns={data.columns[0]:'State',data.columns[3]:'EUR',data.columns[4]:'EE', data.columns[5]:'ELPR', data.columns[6]:'Region'})
```

```python
data.head()
```

Region wise data size

```python
data.groupby('Region').size()
```

**Visualization**

Representing region wise unemployment rate through bar plot




```python
plt.figure(figsize = (10,8))
#c = ['r','y','g','r','k']
plt.bar(data.Region,data.EUR)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate(%)')
```

Representing state wise unemployment rate through horizontal bar chart

```python
plt.figure(figsize = (10,8))
plt.barh(data.State,data.EUR)
plt.xlabel('States')
plt.ylabel('Estimated Unemployment Rate(%)')
```

Representing correalation between features of data through heatmap

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15, 12))
sns.heatmap(data.corr())
```

Representation of Unemployment region wise and state wise through Sun burst

```python
unemploment = data[["State", "Region", "EUR"]]
figure = px.sunburst(unemploment, path=["Region", "State"], 
                     values="EUR", 
                     width=500, height=500, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```
