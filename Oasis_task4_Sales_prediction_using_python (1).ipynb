Task : 

Sales prediction means predicting how much of a product people will buy based on factors
such as the amount you spend to advertise your product, the segment of people you
advertise for, or the platform you are advertising on about your product.
Typically, a product and service-based business always need their Data Scientist to predict
their future sales with every step they take to manipulate the cost of advertising their
product. So let's start the task of sales prediction with machine learning using Python.

Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Loading the dataset into pandas dataframe

```python
data = pd.read_csv('/content/Advertising.csv')
```

```python
#printing first five rows
data.head()
```

```python
data = data.drop(columns = 'Unnamed: 0',axis = 1)
```

No. of rows and columns

```python
data.shape
```

Statistical description of data

```python
data.describe()
```

Visualizing the data

```python
plt.figure(figsize = (8,8))
plt.scatter(data.TV,data.Sales)
plt.title('TV Advertisement Vs Sales')
plt.xlabel('TV Advertisement')
plt.ylabel('Sales')

plt.figure(figsize = (8,8))
plt.scatter(data.Radio,data.Sales)
plt.title('TV Advertisement Vs Sales')
plt.xlabel('TV Advertisement')
plt.ylabel('Sales')

plt.figure(figsize = (8,8))
plt.scatter(data.Newspaper,data.Sales)
plt.title('TV Advertisement Vs Sales')
plt.xlabel('TV Advertisement')
plt.ylabel('Sales')
```

Visualizing the correlation between feature of data through heatmap

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr())
```

Splitting the dataset into features and target

```python
feature = data.drop(columns = 'Sales',axis = 1)
target = data['Sales']
```

```python
print(feature)
```

```python
print(target)
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(feature, target, test_size= 0.2,random_state = 1)
```

```python
print(feature.shape,X_train.shape,X_test.shape)
```

Training the model

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

Model Evaluation - 

Mean Squared Error 

R- squared error

```python
#Prediction for training data
X_train_prediction = model.predict(X_train)
print('Mean Squared Error:',metrics.mean_absolute_error(X_train_prediction,Y_train))
print('R-Squared error',metrics.r2_score(X_train_prediction,Y_train))
```

```python
#Prediction for test data
X_test_prediction = model.predict(X_test)
print('Mean Squared Error:',metrics.mean_absolute_error(X_test_prediction,Y_test))
print('R-Squared error',metrics.r2_score(X_test_prediction,Y_test))
```

Building a Predictive System

```python
input_data = (230.1,37.8,69.2)
#converting the input data to numpy array
input_data_as_np_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_np_array.reshape(1,-1)


prediction = model.predict(input_data_reshaped)
print('Sales:',*prediction)
```
